```python
def repair_null(merged_2):
#Repair null values in dataframe using mode if the column is categorical or mean if the column is numerical.
    for col in merged_2.columns:
        if merged_2[col].dtype == 'object':
            merged_2[col].fillna(merged_2[col].mode()[0], inplace=True)
        else:
            merged_2[col].fillna(merged_2[col].mean(), inplace=True)
        return merged_2
```
